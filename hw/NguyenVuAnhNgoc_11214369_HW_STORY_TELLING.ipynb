```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# covid_avg_deaths.csv

```python
covid_df = pd.read_csv("../data/covid_avg_deaths.csv")
covid_df
```

```python
covid_df['Quarter'] = (covid_df['Month'] - 1) // 3 + 1
quarterly_sum = covid_df.groupby('Quarter')['New_deaths'].sum().reset_index()
print(quarterly_sum)
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=4, ncols=1, figsize=(8, 6))

ax1.plot(covid_df['Month'], covid_df['New_deaths'], color='pink')
ax1.plot(covid_df['Month'].iloc[:3], covid_df['New_deaths'].iloc[:3], color='#c10010', linewidth=2.5)

ax2.plot(covid_df['Month'], covid_df['New_deaths'], color='pink')
ax2.plot(covid_df['Month'].iloc[2:6], covid_df['New_deaths'].iloc[2:6], color='#c10010', linewidth=2.5)

ax3.plot(covid_df['Month'], covid_df['New_deaths'], color='pink')
ax3.plot(covid_df['Month'].iloc[5:10], covid_df['New_deaths'].iloc[5:10], color='#c10010', linewidth=2.5)

ax4.plot(covid_df['Month'], covid_df['New_deaths'], color='pink')
ax4.plot(covid_df['Month'].iloc[9:12], covid_df['New_deaths'].iloc[9:12], color='#c10010', linewidth=2.5)

# Đặt tiêu đề
kws = dict(va='bottom', size=20, weight='bold')
fig.text(0.139, 0.98, 'The virus kills 851 people each day', **kws)
fig.text(0.94, 0.935, 'Average number of daily deadths per month in the US', ha='right', va='bottom', size=17, alpha=0.8)

ax_list = [ax1, ax2, ax3, ax4]
for ax in ax_list:
    ax.spines[['left','right','top','bottom']].set_visible(False)
    ax.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False)
    ax.set_xticklabels([])
    ax.set_yticklabels([])

style = dict(size =  10,color = 'gray',fontstyle='normal')
ax1.axhline(y=1500, xmin=0.5, xmax=0.8, linestyle='-', linewidth=5, alpha=0.3 ,color='grey')
ax1.axhline(y=1500, xmin=0.5, xmax=0.5125, linestyle='-', linewidth=5, alpha=0.3 ,color='red')
ax1.text(0.6, -100, '0', color='grey',size=10)
ax1.text(3.65, 2000, '1,844', **style)
ax1.text(11.7, 2400, '2,247', **style)
ax1.text(7.82, 1650, '2,398', size=10,color = '#c10010', weight='bold')
ax1.text(1.1, -500, 'Jan - Mar', weight='bold', color ='#c10010', rotation=2.5, size=13)


ax2.axhline(y=1500, xmin=0.5, xmax=0.8, linestyle='-', linewidth=5, alpha=0.3 ,color='grey')
ax2.axhline(y=1500, xmin=0.5, xmax=0.63, linestyle='-', linewidth=5, alpha=0.3 ,color='red')
ax2.text(7.65, 1650, '126,203', size=10, color='#c10010', weight='bold')
ax2.text(3.68, 750, 'Mar - Jun', weight='bold', color='#c10010', size=13)


ax3.axhline(y=1500, xmin=0.5, xmax=0.8, linestyle='-', linewidth=5, alpha=0.3 ,color='grey')
ax3.axhline(y=1500, xmin=0.5, xmax=0.73, linestyle='-', linewidth=5, alpha=0.3 ,color='red')
ax3.text(7.65, 1650, '227,178', size=10, color='#c10010', weight='bold')
ax3.text(7.25, 390, 'Jun - Oct', weight='bold', color='#c10010', size=13)


ax4.axhline(y=1500, xmin=0.5, xmax=0.8, linestyle='-', linewidth=5, alpha=0.3 ,color='grey')
ax4.axhline(y=1500, xmin=0.5, xmax=0.8, linestyle='-', linewidth=5, alpha=0.3 ,color='red')
ax4.text(7.65, 1650, '295,406', size=10, color='#c10010', weight='bold')
ax4.text(10.45, 390, 'Oct - Dec', weight='bold', color='#c10010', rotation=32, size=13)

plt.show()
```

# Product_Price.xlsx

```python
product_price_df = pd.read_excel("../data/Product_Price.xlsx")
product_price_df = product_price_df.drop('Unnamed: 1', axis=1)
product_price_df = product_price_df.rename(columns = {'Unnamed: 0': 'Product Name'})
product_price_df
```

Pic 1

```python
fig, ax = plt.subplots(figsize=(10, 5))

# Plot data
ax.plot(product_price_df.set_index('Product Name').T, linewidth=2.5, color='#F1B4BB')
plt.ylim(0, 500)

# Set title
ax.set_title('Average Retail Product Price per Year', size=15, color='black')

# Set yticks
ax.set_yticks([0, 100, 200, 300, 400, 500])
ax.set_yticklabels(['$0', '$100', '$200', '$300', '$400', '$500'])

# Set spines
ax.spines[['top', 'right']].set_visible(False)
ax.spines[['left', 'bottom']].set_color('#DCDCDC')
ax.tick_params(color='#DCDCDC', labelsize=12)            

# Set name lines
kws = dict(color='black', size=14)
ax.text(2007.8, 385, 'A', **kws)
ax.text(2007.8, 350, 'B', **kws)
ax.text(2009.8, 87,  'C', **kws)
ax.text(2010.8, 140, 'D', **kws)
ax.text(2012.8, 90,  'E', **kws)

lines = ax.lines

lines[2].set_markevery([2])         
lines[2].set_marker('o')
lines[2].set_markersize(9)
lines[2].set_markerfacecolor('#DA0C81')

# Highlight
xdata = lines[0].get_xdata()[2:]
ydata = lines[0].get_ydata()[2:]
ax.plot(xdata, ydata, linewidth=2.5, color='#DA0C81')

xdata1 = lines[1].get_xdata()[2:]
ydata1 = lines[1].get_ydata()[2:]
ax.plot(xdata1, ydata1, linewidth=2.5, color='#DA0C81')

plt.show()
```
