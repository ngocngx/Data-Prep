```python
fig, ax1 = plt.subplots()
ax1.bar(['a1','a2','a3'],[10,20,30])
for p in ax1.patches:
    print(type(p))
```

```python
fig, ax1 = plt.subplots()
ax1.bar(['a1','a2','a3'],[10,20,30])
for p in ax1.patches:
    x,y = p.get_xy()
    width = p.get_width()
    height = p.get_height()
    ax1.annotate(xy = (x + width/2, height + 0.5), text = height)
```
