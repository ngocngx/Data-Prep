```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../data/top20_deathtoll.csv')
data.head()
```

```python
data = data.sort_values(by='Total_Deaths', ascending=False)
data = data.reset_index(drop=True)
data = data.head(20)
data
```

```python
data['Country_Other']
```

```python
# Visualize the data
plt.figure(figsize=(10,10))

# horizontal bar chart
plt.barh(data['Country_Other'], data['Total_Deaths'], color='crimson', height=0.5)
plt.gca().invert_yaxis()
plt.title('The Death Toll Worldwide Is 1.5M+', fontsize=20, pad=30, fontweight='bold', loc='left')
# add text Top 20 countries by death toll (Decemeber 2020) as small title

plt.tick_params(axis='x', top=True, labeltop=True, bottom=False, labelbottom=False)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(4.5, 6))
ax.barh(data['Country_Other'], data['Total_Deaths'], color='crimson', height=0.75)

for location in ['left', 'right', 'top', 'bottom']:
    ax.spines[location].set_visible(False)


ax.set_xticks([0, 100_000, 200_000, 300_000])
ax.xaxis.tick_top()
ax

ax.tick_params(axis='x', top=True, labeltop=True, bottom=False, labelbottom=False)
ax.set_title('The Death Toll Worldwide Is 1.5M+', fontsize=10, pad=30, fontweight='bold', loc='left')
ax.invert_yaxis()
ax.text(x=-80000, y=23.5, s='Top 20 countries by death toll (Decemeber 2020)', fontsize=14, fontweight='bold')
plt.show()
```
