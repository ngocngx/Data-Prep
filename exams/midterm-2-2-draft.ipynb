```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# sns.set_style('whitegrid')
sns.set_palette('muted')
pd.options.mode.chained_assignment = None  # default='warn'
```

```python
exchange_rates = pd.read_csv('data-midterm-2-2/euro-daily-hist_1999_2022.csv')
exchange_rates.head()
```

```python
exchange_rates.tail()
```

```python
exchange_rates.info()
```

```python
euro_to_dollar = exchange_rates[['Period\\Unit:', '[US dollar ]']]
euro_to_dollar.columns = ['Time', 'US_dollar']
euro_to_dollar.head()
```

```python
# Time to datetime
euro_to_dollar['Time'] = pd.to_datetime(euro_to_dollar['Time'])

# Sort by time
euro_to_dollar.sort_values('Time', inplace=True)

# US dollar to float coerce
euro_to_dollar['US_dollar'] = pd.to_numeric(euro_to_dollar['US_dollar'], errors='coerce')

euro_to_dollar.info()
```

```python
euro_to_dollar.info()
```

```python
euro_to_dollar['US_dollar'].value_counts(dropna=False)
```

```python
# Fill missing values using backward fill
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].fillna(method='bfill')
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].fillna(method='ffill')
euro_to_dollar.info()
```

```python
# plot the data
plt.plot(euro_to_dollar['Time'], euro_to_dollar['US_dollar'])
plt.show()
```

```python
# Asfreq
euro_to_dollar['Time'].asfreq('D')
euro_to_dollar.head()
```

```python
# Fill null with mean of previous and next values
euro_to_dollar['US_dollar'].fillna(method='ffill', inplace=True)
euro_to_dollar['US_dollar'].fillna(method='bfill', inplace=True)
euro_to_dollar.isnull().sum()
```

```python
# Rolling mean
euro_to_dollar.loc[:, 'rolling_mean'] = euro_to_dollar.loc[:, 'US_dollar'].rolling(30).mean()
euro_to_dollar
```

```python
euro_to_dollar.isnull().sum()
```

```python
# Plot the data
plt.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'])
plt.show()
```

```python
# We show how the euro-dollar rate has changed during the coronavirus pandemic. We can show the 2020 data
# and the 2016-2019 data as a baseline. We can use a line plot.

# 2020 data
euro_to_dollar_2020 = euro_to_dollar[euro_to_dollar['Time'].dt.year == 2020]

# 2016-2019 data
euro_to_dollar_2016_2019 = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2016, 2019)]

# Plot the data
plt.plot(euro_to_dollar_2020['Time'], euro_to_dollar_2020['rolling_mean'], c='red')
plt.plot(euro_to_dollar_2016_2019['Time'], euro_to_dollar_2016_2019['rolling_mean'], c='blue')
plt.show()
```

```python
# We show how the euro-dollar rate changed during the 2007-2008 financial crisis. We can also show the data for
# 2006 and 2009 for comparison. We can use a line plot.

# 2006 data
euro_to_dollar_2006 = euro_to_dollar[euro_to_dollar['Time'].dt.year == 2006]

# 2007-2008 data
euro_to_dollar_2007_2008 = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2007, 2008)]

# 2009 data
euro_to_dollar_2009 = euro_to_dollar[euro_to_dollar['Time'].dt.year == 2009]

# Plot the data
plt.plot(euro_to_dollar_2006['Time'], euro_to_dollar_2006['rolling_mean'], c='blue')
plt.plot(euro_to_dollar_2007_2008['Time'], euro_to_dollar_2007_2008['rolling_mean'], c='red')
plt.plot(euro_to_dollar_2009['Time'], euro_to_dollar_2009['rolling_mean'], c='blue')
plt.show()
```

```python
# We show comparatively how the euro-dollar rate changed under the last three US presidents (George W.Bush
# (2001-2009), Barack Obama (2009-2017), and Donald Trump (2017-2021). We can use a line plot.

# 2001-2008 data
euro_to_dollar_2001_2008 = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2001, 2008)]

# 2009-2016 data
euro_to_dollar_2009_2016 = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2009, 2016)]

# 2017-2021 data
euro_to_dollar_2017_2021 = euro_to_dollar[euro_to_dollar['Time'].dt.year.between(2017, 2021)]

# Plot the data
plt.plot(euro_to_dollar_2001_2008['Time'], euro_to_dollar_2001_2008['rolling_mean'])
plt.plot(euro_to_dollar_2009_2016['Time'], euro_to_dollar_2009_2016['rolling_mean'])
plt.plot(euro_to_dollar_2017_2021['Time'], euro_to_dollar_2017_2021['rolling_mean'])
plt.show()
```

```python
# Clean exchange rates column names
def clean_column_names(df):
    # remove brackets
    df.columns = df.columns.str.strip('[ ]')

    # to camel case
    df.columns = df.columns.str.lower().str.replace(' ', '_')
    
    return df

exchange_rates = clean_column_names(exchange_rates)
exchange_rates.head()
```

```python
exchange_rates.columns
```

```python
exchange_rates.rename(columns={'period\\unit:': 'time'}, inplace=True)
exchange_rates['time'] = pd.to_datetime(exchange_rates['time'])

# To numeric coerce all columns except time
for column in exchange_rates.columns[1:]:
    exchange_rates[column] = pd.to_numeric(exchange_rates[column], errors='coerce')

exchange_rates.info()
```

```python
# Heat map
plt.figure(figsize=(12, 8))
sns.heatmap(exchange_rates.iloc[:, 1:].corr(), cmap='vlag')
plt.show()
```

# Inflation

```python
interest_rates = pd.read_csv('data-midterm-2-2/WB_intereset_rate.csv', skiprows=4)
interest_rates.head()
```

```python
us_interest = interest_rates[interest_rates['Country Name'] == 'United States']
us_interest = us_interest.loc[:, '1999':'2022'].T
us_interest.columns = ['interest_rate']

us_interest['time'] = pd.to_datetime(us_interest.index, format='%Y')
us_interest = us_interest[['time', 'interest_rate']]
us_interest.head()
```

```python
us_inflation = pd.read_excel('data-midterm-2-2/statistic_id191077_us-annual-inflation-rate-1990-2022.xlsx', sheet_name='Data', skiprows=5)
us_inflation = us_inflation.iloc[8:, 1:3]
us_inflation.columns = ['time', 'inflation_rate']
us_inflation['time'] = us_inflation['time'].astype(str).str.zfill(2)
us_inflation['time'] = pd.to_datetime(us_inflation['time'], format='%y')
us_inflation.head()
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax1 = ax.twinx()

ax.plot(us_interest['time'], us_interest['interest_rate'], c='grey', alpha=0.75, linestyle='solid')
ax.plot(us_inflation['time'], us_inflation['inflation_rate'], c='grey', alpha=0.75, linestyle='--')
ax1.plot(euro_to_dollar['Time'], euro_to_dollar['rolling_mean'], c='red', alpha=0.75)

ax.set_xlabel('Time')
ax.set_ylabel('Interest Rate')
ax.set_title('Interest Rate and Inflation Rate in the US')
ax.legend(['Interest Rate', 'Inflation Rate', 'Euro-Dollar Exchange Rate'], loc='upper right')
ax1.legend(['Euro-Dollar Exchange Rate'], loc='upper left')

plt.show()
```

# Continents

```python
# Categorizing the given currencies by continent and storing them in a dictionary

currencies_by_continent = {
    "Oceania": ["australian_dollar", "new_zealand_dollar"],
    "Europe": [
        "bulgarian_lev", "swiss_franc", "cypriot_pound", "czech_koruna", 
        "danish_krone", "estonian_kroon", "uk_pound_sterling", "greek_drachma", 
        "croatian_kuna", "hungarian_forint", "iceland_krona", "lithuanian_litas", 
        "latvian_lats", "maltese_lira", "norwegian_krone", "polish_zloty", 
        "romanian_leu", "russian_rouble", "swedish_krona", "slovenian_tolar", 
        "slovak_koruna", "turkish_lira"
    ],
    "South America": ["brazilian_real"],
    "North America": ["canadian_dollar", "mexican_peso", "us_dollar"],
    "Asia": [
        "chinese_yuan_renminbi", "hong_kong_dollar", "indonesian_rupiah", 
        "israeli_shekel", "indian_rupee", "japanese_yen", "korean_won", 
        "malaysian_ringgit", "philippine_peso", "singapore_dollar", "thai_baht"
    ],
    "Africa": ["south_african_rand"]
}

# Reversing the key-value pairs in the dictionary
# Each currency will now be a key and the continent will be its value

reversed_dict = {}
for continent, currencies in currencies_by_continent.items():
    for currency in currencies:
        reversed_dict[currency] = continent
```

```python
# Plot for each continent

ocenia = exchange_rates[['time', 'australian_dollar', 'new_zealand_dollar']]
ocenia['time'] = pd.to_datetime(ocenia['time'])
ocenia.sort_values('time', inplace=True)

europe = exchange_rates[['time', 'bulgarian_lev', 'swiss_franc', 'cypriot_pound', 'czech_koruna',
         'danish_krone', 'estonian_kroon', 'uk_pound_sterling', 'greek_drachma',
         'croatian_kuna', 'hungarian_forint', 'iceland_krona', 'lithuanian_litas',
         'latvian_lats', 'maltese_lira', 'norwegian_krone', 'polish_zloty',
         'romanian_leu', 'russian_rouble', 'swedish_krona', 'slovenian_tolar',
         'slovak_koruna', 'turkish_lira']]
europe['time'] = pd.to_datetime(europe['time'])
europe.sort_values('time', inplace=True)

south_america = exchange_rates[['time', 'brazilian_real']]
south_america['time'] = pd.to_datetime(south_america['time'])
south_america.sort_values('time', inplace=True)

north_america = exchange_rates[['time', 'canadian_dollar', 'mexican_peso', 'us_dollar']]
north_america['time'] = pd.to_datetime(north_america['time'])
north_america.sort_values('time', inplace=True)

asia = exchange_rates[['time', 'chinese_yuan_renminbi', 'hong_kong_dollar', 'indonesian_rupiah',
            'israeli_shekel', 'indian_rupee', 'japanese_yen', 'korean_won',
            'malaysian_ringgit', 'philippine_peso', 'singapore_dollar', 'thai_baht']]
asia['time'] = pd.to_datetime(asia['time'])
asia.sort_values('time', inplace=True)

africa = exchange_rates[['time', 'south_african_rand']]
africa['time'] = pd.to_datetime(africa['time'])
africa.sort_values('time', inplace=True)
```

```python
# Plot each continent

fig, axes = plt.subplots(3, 2, figsize=(15, 15))
fig.suptitle('Exchange Rates by Continent', fontsize=16)

# Subplot Oceania
for col in ocenia.columns[1:]:
    axes[0, 0].plot(ocenia['time'], ocenia[col], label=col)
    axes[0, 0].set_title('Oceania')
    axes[0, 0].legend()

# Subplot Europe
for col in europe.columns[1:]:
    axes[0, 1].plot(europe['time'], europe[col], label=col)
    axes[0, 1].set_title('Europe')
    axes[0, 1].legend()

# Subplot South America
for col in south_america.columns[1:]:
    axes[1, 0].plot(south_america['time'], south_america[col], label=col)
    axes[1, 0].set_title('South America')
    axes[1, 0].legend()

# Subplot North America
for col in north_america.columns[1:]:
    axes[1, 1].plot(north_america['time'], north_america[col], label=col)
    axes[1, 1].set_title('North America')
    axes[1, 1].legend()

# Subplot Asia
for col in asia.columns[1:]:
    axes[2, 0].plot(asia['time'], asia[col], label=col)
    axes[2, 0].set_title('Asia')
    axes[2, 0].legend()

# Subplot Africa
for col in africa.columns[1:]:
    axes[2, 1].plot(africa['time'], africa[col], label=col)
    axes[2, 1].set_title('Africa')
    axes[2, 1].legend()

plt.show()
```
