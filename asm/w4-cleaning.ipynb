## Q2. 
Đổi tên cột từ định dạng camelcase sang snakecase. Ví dụ: dateCrawled ->date_crawled.
Gợi ý: dùng df.columns

```python
autos_df.columns
```

```python
def camel_to_snake(name):
    name = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', name)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', name).lower()

autos_df.columns = [camel_to_snake(col) for col in autos_df.columns]
autos_df.columns
```

## Q5.
Attribute “model” chứa rất nhiều giá trị phân biệt (characteristics/unique values), tuy nhiên rất nhiều giá trị có số lượng quan sát rất ít. Gom nhóm tất cả các giá trị có số lượng quan sát chiếm tỷ lệ <=0.1% về 1 nhóm gọi tên là OTHER.

```python
# !!!
model_count = autos_df.model.value_counts(normalize=True)
model_count
```

```python
model_count = autos_df['model'].value_counts()
thresh = 0.001 * autos_df.shape[0]

others = model_count[model_count <= thresh]
others
```

```python
# group all models with count less than 0.1% into 'others'
autos_df['model'] = autos_df['model'].apply(lambda x: 'others' if x in others else x)
# autos_df['model'].value_counts()['others']
autos_df['model'].unique()
```

## Q7.
Đối với các features dạng numeric, fillna bởi giá trị median của feature đó. Còn đối với các features dạng categorical thì fillna bởi giá trị quan sát được nhiều nhất.

```python
# date data should be converted to datetime (ordinal data)
autos_df[['year_of_registration', 'date_crawled', 'last_seen']] = autos_df[['year_of_registration', 'date_crawled', 'last_seen']].apply(pd.to_datetime)
# postal code should be converted to object (categorical data)
autos_df['postal_code'] = autos_df['postal_code'].astype('object')
```

```python
autos_df._get_numeric_data()
```

```python
autos_df.info()
```

```python
# Funtion to process price '$5000' to 5000
def process_price(price):
    price = price.replace('$', '')
    price = price.replace(',', '')
    return int(price)

# Funtion to process odometer '100,000 km' to 100000
def process_odometer(odo):
    odo = odo.replace(',', '')
    odo = odo.replace('km', '')
    return int(odo)

autos_df['price'] = autos_df['price'].apply(process_price)
autos_df['price'] = autos_df['price'].astype('int64')
autos_df['odometer'] = autos_df['odometer'].apply(process_odometer)
autos_df['odometer'] = autos_df['odometer'].astype('int64')
autos_df[['price', 'odometer']].head()
```

```python
numeric_col = autos_df.select_dtypes(include=int).columns
category_col = autos_df.select_dtypes(include=object).columns
print('Numeric columns:', numeric_col)
print('Categorical columns:', category_col)
```

```python
autos_df = autos_df.fillna(autos_df[numeric_col].median())
autos_df[numeric_col].isna().sum()
```

```python
autos_df = autos_df.fillna(autos_df[category_col].mode().loc[0])
autos_df[category_col].isna().sum()
```
