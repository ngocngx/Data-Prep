```python
import pandas as pd
import datetime as dt
```

```python
df = pd.read_csv('../data/OnlineRetail2.csv', encoding='latin')
df.head()
```

```python
df.info()
```

```python
# Check distribution of Quantity attributes
df.describe()
```

```python
df.isnull().sum()
```

```python
# Check anomalies
print(len(df[df['Quantity'] < 0]))

# Drop negative quantities, customerID=null, subset 3 columns
df = df[df['Quantity'] > 0].dropna(subset=['CustomerID', 'InvoiceDate', 'InvoiceNo'])
df.shape
```

```python
# Check null values
df.isnull().sum()
```

```python
# Drop duplicates
print(df.shape)
# df.drop_duplicates(inplace=True)
df.drop_duplicates(subset=['InvoiceNo', 'StockCode', 'Quantity'], keep='first', inplace=True)
df.shape
```

```python
df.drop_duplicates(keep='first', inplace=True)
df.shape
```

```python
# Change type of CustomerID to object
df['CustomerID'] = df['CustomerID'].astype('int64').astype('str')
df.head()
```

```python
# Change type of InvoiceDate to datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
df['TotalSales'] = df['Quantity'] * df['UnitPrice']
df.describe()
```

```python
# Create pivot table
df_pivot = pd.DataFrame(columns=['Recency', 'Frequency', 'Monetary'])
df_pivot
```

```python
# Create pivot table to calulate numbers of orders per customer
df_pivot['Frequency'] = df.pivot_table(index='CustomerID', values='InvoiceNo', aggfunc='nunique')
df_pivot

# Last date of purchase each customer
# df_pivot['LastPurchase'] = df.pivot_table(index='CustomerID', values='InvoiceDate', aggfunc='max')
# df_pivot
```

```python
# Calculate total sales each customer
df_pivot['Monetary'] = df.pivot_table(index='CustomerID', values='TotalSales', aggfunc='sum')
df_pivot.head()
```

```python
# Calculate recency each customer
df_pivot['Recency'] = (df.pivot_table(index='CustomerID', values='InvoiceDate', aggfunc='max')['InvoiceDate'] - df['InvoiceDate'].max()).dt.days
df_pivot.head(10)
```

```python
# Qcut for Recency, Frequency, Monetary
df_pivot = df_pivot.apply(lambda x: pd.qcut(x.rank(method='first'), 4, labels=[1, 2, 3, 4]).astype('int64'))
df_pivot.head()
```

```python
df_pivot['TotalScore'] = df_pivot['Recency'] + df_pivot['Frequency'] + df_pivot['Monetary']
df_pivot.head()
```

```python
# Classify customers
df_pivot['Class'] = pd.qcut(df_pivot['TotalScore'], [0, 0.5, 0.8, 1], labels=['low', 'mid', 'high'])
df_pivot.head(10)
```
